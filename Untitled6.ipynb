```python
import imutils
import cv2
from PIL import Image
import requests
import numpy
```

```python
!mkdir tampering
!mkdir tampering/image
```

```python
#here image.open opens the image and get its type and request.get gets the file we requuires from given link
original=Image.open(requests.get('https://www.thestatesman.com/wp-content/uploads/2019/07/pan-card.jpg',stream=True).raw)
tampered=Image.open(requests.get('https://assets1.cleartax-cdn.com/s/img/20170526124335/Pan4.png',stream=True).raw)
```

```python
print(original.size,tampered.size)
print(original.format,tampered.format)
```

```python
#rezing and saving images
original=original.resize((250,160))
original.save('tampering/image/original.png')
tampered=tampered.resize((250,160))
tampered.save('tampering/image/tampered.png')
```

```python
#reading the image using open cv
original=cv2.imread('tampering/image/original.png')
tampered=cv2.imread('tampering/image/tampered.png')
```

```python
#converting the image into gray scale
original_grey=cv2.cvtColor(original,cv2.COLOR_BGR2GRAY)
tampered_grey=cv2.cvtColor(tampered,cv2.COLOR_BGR2GRAY)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(original_grey)
cv2_imshow(tampered_grey)
```

```python
#calculating structural similarity index(ssim) and getting difference image of original and tempered using structural_similarity function
from skimage.metrics import structural_similarity
(score,diff)=structural_similarity(original_grey,tampered_grey,full=True)
diff=(diff*225).astype('uint8')#normalizing the diff and returing it in image format
print(f'the ssimb score is:{score}') 
```

```python
#calculating threesold and thresh
thresh=cv2.threshold(diff,0,225,cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
cnts=cv2.findContours(thresh.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cnts=imutils.grab_contours(cnts)
```

```python
for c in cnts:
  (x,y,w,h)=cv2.boundingRect(c)
  cv2.rectangle(original,(x,y),(x+w,y+h),(0,0,225),2)
  cv2.rectangle(tampered,(x,y),(x+w,y+h),(0,0,225),2)
```
